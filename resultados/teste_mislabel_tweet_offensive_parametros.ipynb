# Teste Mislabel

```python
import pandas as pd

arq = '/content/drive/My Drive/Colab Notebooks/teste_parametros_pka/perc_' +str(perc) + '/offensive/temperatura/base_offensive_temperatura_001.csv'
gptneo = pd.read_csv(arq, sep = ';', decimal = ',')
```

```python
import numpy as np

def prob_softmax(predito):

  mx = np.max(predito, axis=-1, keepdims=True)
  numerator = np.exp(predito - mx)
  denominator = np.sum(numerator, axis=-1, keepdims=True)

  soft = numerator/denominator

  prob = max(soft)

  # classe = np.argmax(soft)

  return prob
```

```python
def calcula_mislabel(base, modelo):
  preds = modelo.predict(base['generated'].to_list())[1]
  base['preds'] = [preds[i] for i in range(preds.shape[0])]
  base['prob_text'] = base['preds'].progress_apply(lambda x:prob_softmax(x))
  base['novo_label'] = base['preds'].progress_apply(lambda x:np.argmax(x))
  base['mislabel'] = (base['novo_label']!= base['new_label']).astype('int')
  return float(base['mislabel'].mean())
```

```python
calcula_mislabel(gptneo, classifica)
```
