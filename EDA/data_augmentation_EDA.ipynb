```python
from textattack.augmentation import Augmenter
```

```python
from textattack.augmentation.recipes import EasyDataAugmenter
augmenter = EasyDataAugmenter(pct_words_to_swap=0.2, transformations_per_example=5)
s = 'What I cannot create, I do not understand.'
augmenter.augment(s)
```

```python
base_cru_1p['generated'] = base_cru_1p['texto'].progress_apply(lambda x : augmenter.augment(x))
```

```python
export = base_cru_1p.copy()
export = export.explode('generated').reset_index(drop=True)
export.head()
```

```python
arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/perc_' +str(perc) + '/emoint/eda/base_emoint_'+str(perc)+'_eda'+str(i)+'.csv'
export.to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```
