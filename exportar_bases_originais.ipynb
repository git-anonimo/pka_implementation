```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
from datasets import load_dataset
from sklearn.model_selection import train_test_split
```

```python
num_gen = 5
perc = 5
```

```python
dataset = load_dataset("emotion")
X = dataset['train']['text']
y = dataset['train']['label']

data = {'text':X, 'label': y}
train = pd.DataFrame.from_dict(data)

data = {'text': dataset['test']['text'], 'label': dataset['test']['label']}
test = pd.DataFrame.from_dict(data)

arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_emotions_test.csv'
test[['text', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')

arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_emotions_train.csv'
train[['text', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```

```python
arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_emotions_train.csv'
train= pd.read_csv(arquivo, encoding='utf-8', sep=';', decimal = ',')

X= train['text'].values
y= train['label'].values

for i in range(num_gen):


  X_resto, X_treino_1p, y_resto, y_treino_1p = train_test_split(X, y, stratify=y, test_size=(perc/100), random_state=i)

  data = {'texto': X_treino_1p, 'label': y_treino_1p}
  base_cru_1p = pd.DataFrame.from_dict(data)
  base_cru_1p.shape

  arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/perc_' +str(perc) + '/emotion/base_emotions_'+str(perc)+'_original'+str(i+1)+'.csv'
  base_cru_1p[['texto', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```

# SST

```python
import numpy as np
dataset = load_dataset("sst")
X = dataset['train']['sentence']
y = dataset['train']['label']

data = {'text':X, 'label': y}
train = pd.DataFrame.from_dict(data)
train['label'] = np.where(train['label']>0.5, 1, 0)

data = {'text': dataset['test']['sentence'], 'label': dataset['test']['label']}
test = pd.DataFrame.from_dict(data)
test['label'] = np.where(test['label']>0.5, 1, 0)

arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_sst_test.csv'
test[['text', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')

arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_sst_train.csv'
train[['text', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```

```python
arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/original/base_sst_train.csv'
train= pd.read_csv(arquivo, encoding='utf-8', sep=';', decimal = ',')

X= train['text'].values
y= train['label'].values

for i in range(num_gen):


  X_resto, X_treino_1p, y_resto, y_treino_1p = train_test_split(train['text'],train['label'], stratify=train['label'], test_size=(perc/100), random_state=i)

  data = {'texto': X_treino_1p, 'label': y_treino_1p}
  base_cru_1p = pd.DataFrame.from_dict(data)
  base_cru_1p['label'] = np.where(base_cru_1p['label']>0.5, 1, 0)
  base_cru_1p.shape

  arquivo = '/content/drive/My Drive/Colab Notebooks/t5_key_text/perc_' +str(perc) + '/sst/base_sst_'+str(perc)+'_original'+str(i+1)+'.csv'
  base_cru_1p[['texto', 'label']].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```
